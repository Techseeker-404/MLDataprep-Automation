# Dropping down Entire columns which contain more than 45% of null values than the length of DataFrame. 

```python
def drop_overna(data, perc):
    for cols in data.columns:
        if data[cols].isnull().sum() > int(perc * len(df)):
                data=data.drop([cols],axis=1)
        else:
               pass
    return data
data = drop_overna(df,0.45)
df=data      
df
```

```python
df.info()
```

# Create a For loop to fill rest of the null values with mean if the column values are numeric and with mode if the columns datatype is object.

```python
for cols in df.columns:
    if df[cols].isnull:
        if df[cols].dtype == object:
            df[cols] = df[cols].fillna(df[cols].mode()[0])
        else:
            df[cols] = df[cols].fillna(df[cols].mean())
```

# Most rlevant part of numeric selection as this will declare those features which are having a Correlation value less than 0.45 with target .

```python
# create a list to select out those filtered varible features which are more than 0.45 correlation value with target.
filteredSalePriceCorr = []
def filterCor(dafcol,bound):
    for cor in dafcol:
        if cor >= bound:
            filteredSalePriceCorr.append(True)
        else:
            filteredSalePriceCorr.append(False)
    return filteredSalePriceCorr
filterCor(corr.SalePrice,0.5)
```
